### Wiederholung reduzieren mit Funktionen

```python
turtle.bye()
```

```python
turtle.getscreen()
stift = turtle.Turtle()
```

```python
def schritte_vor():
    stift.right(90)
    stift.forward(100)
    stift.forward(100)
```

```python
def ausfuellen_rot():
    stift.fillcolor("red")
    stift.begin_fill()
    for _ in range(4):
        stift.forward(100)
        stift.right(90)
    stift.end_fill()
```

```python
def ausfuellen_gelb():
    stift.fillcolor("yellow")
    stift.begin_fill()
    for _ in range(4):
        stift.forward(100)
        stift.right(90)
    stift.end_fill()
```

```python
def ausfuellen_gruen():
    stift.fillcolor("green")
    stift.begin_fill()
    for _ in range(4):
        stift.forward(100)
        stift.right(90)
    stift.end_fill()
```

```python
schritte_vor()
ausfuellen_rot()
schritte_vor()
ausfuellen_gelb()
schritte_vor()
ausfuellen_gruen()
```

### Es gibt immer noch ein Problem ...

**Frage:** Was müsste man tun, um die Quadratgröße von 100 auf 110 zu ändern?

### Noch weniger Wiederholung mit Funktionen und Parametern

```python
def farbig_ausfuellen(farbe):
    stift.fillcolor(farbe)
    stift.begin_fill()
    for _ in range(4):
        stift.forward(100)
        stift.right(90)
    stift.end_fill()
```

```python
schritte_vor()
farbig_ausfuellen("red")
schritte_vor()
farbig_ausfuellen("yellow")
schritte_vor()
farbig_ausfuellen("green")
```

**Frage**: Und was muss man jetzt nur noch machen um die Quadratgröße von 100 auf 110 zu ändern?

# Sort images in Dictionary

```python
import os
import pathlib
import collections
import datetime
```

```python
file_mapping = collections.defaultdict(list)
path_ = "/Volumes/Macintosh HD/jDownloader/Bilder/Handy_Fotos"
path_lib = pathlib.Path(path_)

for file_ in path_lib.iterdir():
    pathlib_file = pathlib.Path(file_)
    if pathlib_file.suffix.lower() in ['.jpg', '.jpeg']:
        creation_timestamp = pathlib_file.stat().st_mtime
        creation_datetime = datetime.datetime.fromtimestamp(creation_timestamp)
        date = creation_datetime.strftime("%Y-%m-%d")
        file_mapping[date].append(pathlib_file)
```
